```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_excel('/content/newcar.xls',sheet_name=list(range(0,6)),header=3)
```

```python
for n in range (0,6):
  #rename column
    df[n].rename(columns={'ยี่ห้อรถ':'brand',
                    'รุ่นรถ':'model',
                   'จำนวนรถ':'Number_of_Newcar'},inplace=True)
    #create column type_reg
    df[n]['Type_register']=df[n].apply(lambda x:x['brand'] if np.isnan(x['รวมยี่ห้อ']) else np.nan,axis=1)
      #add fillna column type and brand
    df[n]['Type_register']=df[n]['Type_register'].fillna(method='pad')
    df[n]['brand']=df[n]['brand'].fillna(method='pad')
    #drop column Nan
    df[n]=df[n].dropna(subset=['model']).reset_index(drop=True)
    #change type odj to int
    df[n]['Number_of_Newcar']=df[n]['Number_of_Newcar'].astype(int)
    df[n].set_index('brand', inplace=True)
    df[n].drop(['รวมยี่ห้อ'],axis=1,inplace=True)
    #df[n]['Year'] =df[n].apply(lambda x: '2024' if pd.notna(x['model']) else pd.isna(x['model']), axis=1)
    #create column date
    df[n]['DATE']=df[n].apply(lambda x:'2024'+'-'+ str(n+1) if pd.notna(x['model']) else pd.isna(x['model']), axis=1)
    df[n]['DATE']=pd.to_datetime(df[n]['DATE'] )










    #df[n]['Month'] = df[n]['Month'].month_name()



    #df[n]['DATE']=df[n]['Year'].astype(str)+'-'+df[n]['Month'].astype(str)

    #df[n]['DATE']=pd.to_datetime(df[n]['DATE'] )
    #df[n].drop(['Year','Month'],axis=1,inplace=True)


```
